### 4. initial column (or row) clean-up

```python
#Finds columns with only 1 unique value
for col in rsm:
    unique = rsm[col].unique()
    if len(unique) == 1:
        print(col, 'HAS ONLY ONE UNIQUE VALUE, ', unique)
    else:
        print(col, 'has multiple unique values')
        
#Drop the two columns with unique values      
rsm.drop(['BOROUGH','EASE-MENT'], axis = 1)
```

### 6. describe the kind of buildings that were sold

```python
#Import necessary plotting module
import matplotlib.pyplot as plt

#Make separate dataframes with only sales of each category
onefam = rsm[rsm['Building Class Category'].str.contains("ONE FAMILY DWELLINGS")].index
office = rsm[rsm['Building Class Category'].str.contains("OFFICE")].index
condo = rsm[rsm['Building Class Category'].str.contains("CONDO")].index

#Used to calculate the "other" by being subtracted from
total = rsm['Building Class Category'].index

#Plot a bar graph where x-value is the mentioned categories and y-value is the number of sales, which can be found from length of category array
plt.bar(x=['One Family Homes', 'Office Buildings', 'Condominiums', 'Other'],height = [len(onefam), len(office), len(condo), len(total)-len(onefam)-len(office)-len(condo)])
```

### 7. calculate summary statistics for the prices of properties sold for all of Manhattan and for a couple of select neighborhood

```python
#Manhattan Summary Statistics formatted to not have scientific notation
rsm[' Sale Price '].describe().apply(lambda x: format(x, 'f'))
```

```python
#Alphabet City
#Use loc to filter only specific neighborhood and set column as 'Sale Price' since thats all we want to know the statistics of
alph = rsm.loc[rsm['Neighborhood'] == 'ALPHABET CITY'][' Sale Price ']
alph.describe().apply(lambda x: format(x, 'f'))
```

```python
#Washington Heights Upper

wash = rsm.loc[rsm['Neighborhood'] == 'WASHINGTON HEIGHTS UPPER'][' Sale Price ']
wash.describe().apply(lambda x: format(x, 'f'))
```

Manhattan overall has a higher mean than the two neighborhoods

Alphabet City has incredibly low purchases, with the 25th percentile sale being only 2500 as oppsed to Manhattan 300000, Washington Heights Upper 355000
