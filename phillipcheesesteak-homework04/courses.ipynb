## 1. Read the 2021 course schedule into a DataFrame

```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Read the whole html
cs2021 = BeautifulSoup(open("spr2021schedule.htm"), 'html.parser')
#The schedule information are all inside .schedule-listing class
schedulelisting = cs2021.select('.schedule-listing')[0]
```

#### Parse number-section into array

```python
#Count number of 'li', which are lists of individual classes
numseccount = schedulelisting.select('li')
numbersectionarr = []
#Append all numbersection into array
for i in range(len(numseccount)):
    numbersectionarr.append(schedulelisting.select('li')[i]['id'])
#Clean into appropriate format
for i in range(len(numbersectionarr)):
    numbersectionarr[i] = numbersectionarr[i].upper()
```

#### Parse course name into array

```python
#All course names are inside class .expand
coursename = schedulelisting.select('.expand')
#Array to store course names
coursenamearr = []
#Append while removing newlines
for i in range(len(coursename)):
    coursenamearr.append(coursename[i].get_text())
    coursenamearr[i] = " ".join(coursenamearr[i].strip().split())
```

#### Parse instructor name and time into array

```python
#Since instructor and time columns share similar attributes with class col-sm 2 in HTML, parse and clean them simultaneously
instructorAndTime = schedulelisting.select('.col-sm-2')
instructorAndTimeArr = []
for i in range(len(instructorAndTime)):
    #Append all text elemeents into array
    instructorAndTimeArr.append(instructorAndTime[i].get_text())
    #Delete new lines and spaces
    instructorAndTimeArr[i] = " ".join(instructorAndTimeArr[i].strip().split())
    #Rid of "Office Hours"
    instructorAndTimeArr[i] = instructorAndTimeArr[i].replace(" Office Hours", "")
    
timeArr = []
instructorArr = []
for i in range(0,len(instructorAndTimeArr),3):
    instructorArr.append(instructorAndTimeArr[i])
    timeArr.append(instructorAndTimeArr[i+1])
```

#### Read the course schedule into a Dataframe

```python
df = pd.DataFrame()
df['Number-Section'] = numbersectionarr
df['Name'] = coursenamearr
df['Instructor'] = instructorArr
df['Time'] = timeArr
df
```

#### Break apart Number-Section into Number and Section

```python
df['Number'] = df['Number-Section'].str.extract('(.*-.*)-')
df['Section'] = df['Number-Section'].str.extract('.*-.*-(\d*)')
df = df.drop(['Number-Section'],axis = 1)
```

## 2. Read the 2021 course catalog into a Dataframe

```python
#Read the whole html
coursecat = BeautifulSoup(open("coursecat.htm", encoding = 'utf-8'), 'html.parser')
#The schedule information are all inside .schedule-listing class
coursecat1 = coursecat.select('.courses-listing')[0]
```

#### Parse number into array

```python
#Count number of 'li', which are lists of individual classes
numcount = coursecat1.select('li')
numberarr = []
#Append all numbersection into array
for i in range(len(numcount)):
    numberarr.append(coursecat1.select('li')[i]['id'])
#Clean into appropriate format
for i in range(len(numberarr)):
    numberarr[i] = numberarr[i].upper()
```

#### Parse Prerequesites into array

```python
#All prerequisites are inside class .expandable
prerequisites = coursecat1.select('.bold')

#Array to store prerequisites
prereqarr = []
counter = 0
#Append while removing newlines, space
for i in range(2, len(prerequisites), 3):
    prereqarr.append(prerequisites[i].get_text())
    prereqarr[counter] = " ".join(prereqarr[counter].strip().split())
    counter += 1
```

#### Parse Points to array

```python
pointarr = []
counter = 0
for i in range(1, len(prerequisites), 3):
    pointarr.append(prerequisites[i].get_text())
    pointarr[counter] = pointarr[counter].strip().split()[0]
    counter += 1
```

#### Read arrays into dataframe

```python
df1 = pd.DataFrame(columns = ['Number', 'Prereqs', 'Points'])
df1['Number'] = numberarr
df1['Prereqs'] = prereqarr
df1['Points'] = pointarr
```

## 3. Put together both DataFrames

```python
df2 = pd.merge(df,df1, on='Number', how='left')
pd.set_option('display.max_rows', 200)
df2
```
